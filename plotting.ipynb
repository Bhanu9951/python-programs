```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 10,15])
ypoints = np.array([0, 450,400])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([8, 10, 30, 60,70])
mylabels=["sea water","drinking water","daily purpose use","rivers","drainage water"]

plt.pie(y, labels = mylabels)
plt.legend(title ="water")
plt.show() 
```
